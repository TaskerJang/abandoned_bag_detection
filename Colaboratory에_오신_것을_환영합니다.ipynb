```
import cv2

# 웹캠을 열어서 캡처 객체 생성
cap = cv2.VideoCapture(0)

# 카메라가 정상적으로 열렸는지 확인
if not cap.isOpened():
    print("카메라를 열 수 없습니다.")
else:
    print("카메라를 열었습니다.")

# 카메라에서 프레임을 읽어오고 화면에 표시
while True:
    ret, frame = cap.read()
    if not ret:
        break

    cv2.imshow("Camera", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 사용이 끝나면 캡처 객체와 창을 닫음
cap.release()
cv2.destroyAllWindows()
```
